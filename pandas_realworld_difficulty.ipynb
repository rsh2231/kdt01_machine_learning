### Î¨∏Ï†ú 5
Age Ïª¨ÎüºÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Íµ¨Í∞ÑÏùÑ ÎÇòÎàÑÍ≥† Í∑∏Î£πÎ≥Ñ ÌèâÍ∑†ÏùÑ Íµ¨ÌïòÏÑ∏Ïöî.

```python
import pandas as pd

# ÏòàÏãú: titanic.csv ÌååÏùº Î∂àÎü¨Ïò§Í∏∞
# df = pd.read_csv('titanic.csv')
# Ïó¨Í∏∞Ïóê ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏÑ∏Ïöî

# Íµ¨Í∞Ñ ÏÑ§Ï†ï: 0-18, 19-35, 36-60, 61 Ïù¥ÏÉÅ
bins = [0, 18, 35, 60, float('inf')]
labels = ['0-18', '19-35', '36-60', '61+']

# Íµ¨Í∞ÑÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÎÇòÎàÑÍ∏∞
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Í∞Å Í∑∏Î£πÎ≥Ñ Age ÌèâÍ∑† Íµ¨ÌïòÍ∏∞
avg_age_by_group = df.groupby('Age_group')['Age'].mean()
print(avg_age_by_group)
df
```
