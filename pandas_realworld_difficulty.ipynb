### 문제 5
Age 컬럼을 기준으로 구간을 나누고 그룹별 평균을 구하세요.

```python
import pandas as pd

# 예시: titanic.csv 파일 불러오기
# df = pd.read_csv('titanic.csv')
# 여기에 코드를 작성하세요

# 구간 설정: 0-18, 19-35, 36-60, 61 이상
bins = [0, 18, 35, 60, float('inf')]
labels = ['0-18', '19-35', '36-60', '61+']

# 구간을 기준으로 나누기
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# 각 그룹별 Age 평균 구하기
avg_age_by_group = df.groupby('Age_group')['Age'].mean()
print(avg_age_by_group)
df
```
